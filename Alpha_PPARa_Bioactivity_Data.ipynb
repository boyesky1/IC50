```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('PPARalpha')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

Selecting EC50 values reported in nM units

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="Activity")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(3)
```

```python
df.standard_type.unique()
```

```python
df.to_csv('PPARa_bioactivity_data_nM.csv', index=False)
```

Deleting missing data

```python
df2 = df[df.value.notna()]
df2
```

Turning data in the "Value" column to float type

```python
df2['value']=pd.to_numeric(df2['value'], downcast='float')
```

```python
print(df2['value'].dtypes)
```

Converting nM to uM

```python
x=float(0.001)
df2['value']=x*df2['value']
df2['value']
```

```python
df2.to_csv('PPARa_bioactivity_data_no_missing_data_values .csv', index=False)
```

Selecting EC50 values reported in uM units




```python
activity = new_client.activity
res1 = activity.filter(target_chembl_id=selected_target).filter(standard_type="Activity",)
```

```python
df3 = pd.DataFrame.from_dict(res1)
```

```python
df4 = df3[df3.standard_value.notna()]
```

```python
df4.to_csv('PPARa_bioactivity_data_uM.csv', index=False)
```

Merging the two CSV files in one

```python
csv1 = pd.read_csv("PPARa_bioactivity_data_uM.csv")
csv2 = pd.read_csv("PPARa_bioactivity_data_no_missing_data_values .csv")
concate_data = pd.concat([csv1,csv2])
df5=concate_data
df5.to_csv('PPARa_bioactivity_data_all_uM.csv', index=False)
df5
```

Data preprocessing

```python
bioactivity_class = []
for i in df5.value:
  if float(i) >= 10:
    bioactivity_class.append("Partial Agonist")
  else:
    bioactivity_class.append("Full Agonist")
```


Iterate the molecule_chembl_id to a list

```python
mol_cid = []
for i in df5.molecule_chembl_id:
  mol_cid.append(i)
```

Iterate canonical_smiles to a list

```python
canonical_smiles = []
for i in df5.canonical_smiles:
  canonical_smiles.append(i)
```

Iterate standard_value to a list

```python
value = []
for i in df5.value:
  value.append(i)
```


Combine the 4 lists into a dataframe

```python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, value))
df6 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'value_uM'])
```
